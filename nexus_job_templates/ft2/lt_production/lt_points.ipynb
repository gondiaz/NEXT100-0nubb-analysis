```python
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
rmax = 492
n  = 16
dr = 2
    
#-----------------------------
#----- Create grid points ----
#-----------------------------
X = np.round(np.linspace(-(rmax-dr), (rmax-dr), n), 2)
x, y = np.meshgrid(X, X)
r = (x**2 + y**2)**0.5
sel = r<rmax
x, y = x[sel], y[sel]
```

```python
fig = plt.figure(figsize=[5,5])
ax = fig.add_subplot(111)

ax.scatter(x, y, s=2, c="r")

theta=np.linspace(0, 2*np.pi, 200)
ax.plot(rmax*np.cos(theta), rmax*np.sin(theta))
```
